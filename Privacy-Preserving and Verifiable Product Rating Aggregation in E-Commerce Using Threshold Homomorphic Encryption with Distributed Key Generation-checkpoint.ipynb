```python
import time, random

def bulletproof_simulator():
    # Simulated Bayer–Groth proof latency: 20–50 ms
    delay = random.uniform(0.020, 0.050)
    t1 = time.perf_counter()
    time.sleep(delay)
    t2 = time.perf_counter()

    # Simulated proof size: 1.5–2.5 KB
    size_bytes = random.randint(1500, 2500)

    return (t2 - t1) * 1000, size_bytes / 1024  # ms, KB

# Test
bulletproof_simulator()

```

```python
def shamir_split(secret, n=7, t=3, prime=2**256 - 189):
    coeffs = [secret] + [random.randint(1, prime-1) for _ in range(t-1)]
    shares = []
    for i in range(1, n+1):
        x = i
        y = sum([coeffs[j] * pow(x, j, prime) for j in range(t)]) % prime
        shares.append((x, y))
    return shares

def benchmark_shamir():
    secret = random.getrandbits(256)
    t1 = time.perf_counter()
    _ = shamir_split(secret)
    t2 = time.perf_counter()
    return (t2 - t1) * 1000  # ms

benchmark_shamir()

```

```python
def benchmark_paillier_repeated(R=50):
    times=[]
    for _ in range(R):
        t1=time.perf_counter()
        public_key.encrypt(random.randint(1,5))
        t2=time.perf_counter()
        times.append((t2-t1)*1000)
    return {
        "min_ms": min(times),
        "max_ms": max(times),
        "avg_ms": sum(times)/len(times)
    }

benchmark_paillier_repeated(50)

```

```python
from phe import paillier
import time, random
import numpy as np

# Generate keypair (2048-bit)
public_key, private_key = paillier.generate_paillier_keypair(n_length=2048)

def benchmark_paillier(batch=1000):
    """Benchmark Paillier encryption for a batch of ratings."""
    times = []
    for _ in range(batch):
        r = random.randint(1,5)
        t1 = time.perf_counter()
        public_key.encrypt(r)
        t2 = time.perf_counter()
        times.append((t2 - t1) * 1000)  # convert to ms
    
    return {
        "min_ms": np.min(times),
        "max_ms": np.max(times),
        "avg_ms": np.mean(times),
        "batch": batch
    }

result = benchmark_paillier(batch=1000)
result
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# داده‌های شما
protocols = ['Hybrid HE + DKG', 'Malicious-Secure MPC', 'Blockchain-based Protocols']
times = [0.230446900000004, 4.5773841000000175, 3.4424730999999724]  # زمان برای 2000 نمونه
times_500k = [57.611725000001, 1144.3460250000044, 860.6182749999931]  # زمان برای 500,000 رأی

# ساختار داده‌ها برای x و y
x = np.array([0, 1, 2])  # نماینده پروتکل‌ها
y = np.array([0, 1])  # نماینده زمان‌ها (برای زمان‌های نمونه و 500k)

# ایجاد شبکه برای x و y
x_mesh, y_mesh = np.meshgrid(x, y)

# ایجاد داده‌های z برای نمایش در نمودار
z = np.array([[times[i], times_500k[i]] for i in range(3)])

# اصلاح ابعاد z به گونه‌ای که برای plot_surface مناسب باشد
z = np.transpose(z)  # معکوس کردن ابعاد z

# رسم نمودار سه‌بعدی
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# رسم سطح
surf = ax.plot_surface(x_mesh, y_mesh, z, cmap='hot', edgecolor='none')

# برچسب‌ها
ax.set_xlabel('Protocols', labelpad=15)  # افزایش فاصله برچسب x
ax.set_ylabel('Time (for 2000 samples vs. 500k votes)', labelpad=15)  # افزایش فاصله برچسب y
ax.set_zlabel('Time (seconds)', labelpad=10)  # افزایش فاصله برچسب z

# نمایش رنگ‌بندی و انتقال به سمت راست
cbar = fig.colorbar(surf, orientation='vertical', pad=0.1)
cbar.set_label('Time (seconds)', labelpad=10)

# تنظیمات برای نمایش نمودار مشابه به آنچه که خواسته‌اید
plt.xticks(x, protocols, rotation=30, ha='right')  # چرخاندن نوشته‌های محور x
plt.yticks(y, ['2000 Samples', '500,000 Votes'], rotation=0)  # نوشته‌های محور y
plt.tight_layout()  # برای جلوگیری از تداخل نوشته‌ها

plt.show()
plt.savefig("comparison_protocols_plot_blue.pdf", format='pdf')
```
